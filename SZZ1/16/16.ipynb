```python
import numpy as np
import matplotlib.pyplot as plt

# Definice kvantity Q
Q = np.linspace(0.1, 20, 400)  # Pedpokldme, e Q > 0 pro vpoty prmrnch funkc

# Celkov pjmy a nklady
TR = -2*Q**3 + 96*Q**2 + 198*Q
TC = Q**3 - 60*Q**2 + 1300*Q + 4096

# Mezn pjmy a nklady
MR = np.gradient(TR, Q)
MC = np.gradient(TC, Q)

# Prmrn pjmy a nklady
AR = TR / Q
AC = TC / Q

# Vykreslen graf
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Graf pro celkov pjmy a nklady
ax[0].plot(Q, TR, label='Celkov pjmy TR(Q)', color='blue')
ax[0].plot(Q, TC, label='Celkov nklady TC(Q)', color='red')
ax[0].set_title('Celkov pjmy a nklady')
ax[0].set_xlabel('Mnostv Q')
ax[0].set_ylabel('Hodnoty TR a TC')
ax[0].legend()

# Graf pro prmrn a mezn pjmy a nklady
ax[1].plot(Q, AR, label='Prmrn pjmy AR(Q)', color='blue', linestyle='dashed')
ax[1].plot(Q, AC, label='Prmrn nklady AC(Q)', color='red', linestyle='dashed')
ax[1].plot(Q, MR, label='Mezn pjmy MR(Q)', color='blue')
ax[1].plot(Q, MC, label='Mezn nklady MC(Q)', color='red')
ax[1].set_title('Prmrn a Mezn pjmy a Nklady')
ax[1].set_xlabel('Mnostv Q')
ax[1].set_ylabel('Hodnoty AR, AC, MR, MC')
ax[1].legend()

plt.tight_layout()
plt.show()
```
