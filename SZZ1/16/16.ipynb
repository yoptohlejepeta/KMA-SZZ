```python
import numpy as np
import matplotlib.pyplot as plt

# Definice kvantity Q
Q = np.linspace(0.1, 20, 400)  # Předpokládáme, že Q > 0 pro výpočty průměrných funkcí

# Celkové příjmy a náklady
TR = -2*Q**3 + 96*Q**2 + 198*Q
TC = Q**3 - 60*Q**2 + 1300*Q + 4096

# Mezní příjmy a náklady
MR = np.gradient(TR, Q)
MC = np.gradient(TC, Q)

# Průměrné příjmy a náklady
AR = TR / Q
AC = TC / Q

# Vykreslení grafů
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Graf pro celkové příjmy a náklady
ax[0].plot(Q, TR, label='Celkové příjmy TR(Q)', color='blue')
ax[0].plot(Q, TC, label='Celkové náklady TC(Q)', color='red')
ax[0].set_title('Celkové příjmy a náklady')
ax[0].set_xlabel('Množství Q')
ax[0].set_ylabel('Hodnoty TR a TC')
ax[0].legend()

# Graf pro průměrné a mezní příjmy a náklady
ax[1].plot(Q, AR, label='Průměrné příjmy AR(Q)', color='blue', linestyle='dashed')
ax[1].plot(Q, AC, label='Průměrné náklady AC(Q)', color='red', linestyle='dashed')
ax[1].plot(Q, MR, label='Mezní příjmy MR(Q)', color='blue')
ax[1].plot(Q, MC, label='Mezní náklady MC(Q)', color='red')
ax[1].set_title('Průměrné a Mezní příjmy a Náklady')
ax[1].set_xlabel('Množství Q')
ax[1].set_ylabel('Hodnoty AR, AC, MR, MC')
ax[1].legend()

plt.tight_layout()
plt.show()
```
